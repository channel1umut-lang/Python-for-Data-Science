```python
#multi index
```

```python
first_index = ["Simpson","Simpson","Simpson","South Park","South Park","South Park"]
```

```python
inner_index = ["Homer","Bart","Marge","Cartman","Kenny","Kyle"]
```

```python
zipped_index = list(zip(first_index,inner_index))
```

```python
zipped_index
```

```python
zipped_index = pd.MultiIndex.from_tuples(zipped_index) # birden fazla index varsa bunu getirmek icin
```

```python
zipped_index
```

```python
sample_values = np.ones((6,2))
```

```python
big_df = pd.DataFrame(sample_values, index=zipped_index, columns=["Maas","Yas"])
```

```python
big_df
```

```python
big_df["Maas"]
```

```python
#big_df.loc["Homer"] does not work because it is multiple index
big_df.loc["Simpson"]
```
