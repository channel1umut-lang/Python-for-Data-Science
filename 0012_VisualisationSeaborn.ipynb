```python
plt.scatter("Height", "Weight", data=df) # noktagrafigi
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #better visualisation
sns.scatterplot(x="Height", y="Weight", data=df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #better visualisation
sns.scatterplot(x="Height", y="Weight", hue="Sex", data=df) # hue =  cinsiyete gore boy ve kilo dagilimini gormek icin hue kullaniyoruz
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #better visualisation
sns.scatterplot(x="Height", y="Weight", hue="Medal", data=df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #better visualisation
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", data=df) #style = kendisi bazi sekiller atiyor
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") #better visualisation
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", size = "Age", data=df) #age = greater data point greater age does itself
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight", hue = "Sex", data=df)#line graph
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("white") #renkleri ayarlio
sns.displot(x="Height", hue="Sex", data=df)
plt.ylabel("Frequency")
plt.title("Athletes High distrubition")
plt.show()
```

```python
sns.set_style("white") #renkleri ayarlio
sns.displot(x="Height", hue="Sex", data=df, kind="kde")# kde = cizgi yapti
plt.ylabel("Frequency")
plt.title("Athletes High distrubition")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=df) #bar graph
plt.title("Medals by height")
plt.show()
```
