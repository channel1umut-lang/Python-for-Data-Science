```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#box plot, cikinti verileri gozlemlemek icin iyi bir yontem
```

```python
#median, quartile (25%, 50%, 75%), min max 1.5 * IQR, Outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(10,5)) # figure size = resmin eni boyu
sns.boxplot(y=data)
plt.title('Box Plot')
plt.ylabel('Data Values')
#plt.grid(True) #izgaralar
plt.show()
#Dikdortgenin ortasindaki cizgi medyani, dikdortgenin kendisi datalarin yogun olarak dagildigi yeri gosteriyor
#Dikdortgenin ustundeki T datanin nereye kadar ciktigini yani 75%*1.5, altindaki T de nereye kadar indigini gosteriyor yani 25%*1.5.
#Dokdorgenin ust cizgisi 75%ligi alt cizgisi 25% ligi gosteriyor
#ustteki yuvarlak ise outlier i gosteriyor.

```

```python
df = sns.load_dataset('titanic')
df
```

```python
plt.figure(figsize=(13,8))

plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data  = df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data  = df)
plt.title("fare by class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout() # daha duzgun gostermesi icin bu kodu kullanabiliriz
plt.show()
```
