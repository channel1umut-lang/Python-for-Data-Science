```python
df_merge1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/7-merge_data1.csv")
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSets/7-merge_data2.csv")
```

```python
df_merge2
```

```python
#marge - ourter join
```

```python
df_merged_outer = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="outer")
```

```python
df_merged_outer
```

```python
#merge - left join
```

```python
df_merged_left = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "left") #df_merge1 left deki tablo oluyor bu satirda sol tarafta yazilmis cunku
```

```python
df_merged_left
```

```python
df_merged_right = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "right")
```

```python
df_merged_right
```

```python
#merge inner
```

```python
df_merge = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "inner")
```
