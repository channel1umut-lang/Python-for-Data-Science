```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 **2
```

```python
data2
```

```python
plt.subplots() # bunu yapinca otomatik bir figur bir axis donduruyor.
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2, color="#1E0703", alpha = 0.5)# colors hexa code,,, alpha transparanlastiriyor
my_axes.plot(data2,data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1,data1 +2, color = "blue", linewidth = 2)
new_axes.plot(data1,data1 +3, color = "yellow", linewidth = 4)
new_axes.plot(data1,data1 +4, color = "red", linestyle = "-.")
new_axes.plot(data1,data1 +5, color = "red", linestyle = "--")
new_axes.plot(data1,data1 +6, color = "red", linestyle = ":")
new_axes.plot(data1,data1 +7, color = "black", linestyle = ":", marker = "o", markersize = 8, markerfacecolor = "blue")
new_axes.plot(data1,data1 +8, color = "black", linestyle = "-.", marker = "+", markersize = 10)

plt.show()
```
