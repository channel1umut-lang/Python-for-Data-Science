```python
def salary_catagory(salary):
  if salary<50000:
    return "Low"
  elif 50000<=salary<100000:
    return "Medium"
  else:
    return "High"
```

```python
df["Salary"].apply(salary_catagory)
```

```python
df["Salary_Catagory"] = df["Salary"].apply(salary_catagory) #added to main df
```

```python
#if experience > 10 ---> Performance_Score +1 bonus example
```

```python
def new(Experience):
  if Experience>10:
    return 1
  else:
    return 0
```

```python
df["Experience"].apply(new)
```

```python
df["new_experience"] = df["Experience"].apply(new)
```

```python
df
```

```python
df["New_Performance"] = df["new_experience"] + df["Performance_Score"]
```

```python
def adjust_new(row):
  if row["Experience"] > 10:
    return row["Performance_Score"] + 1
  else:
    return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(adjust_new, axis=1)
```
