```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
len(weight_list)
```

```python
plt.plot(age_list, weight_list, "r") # r--->red y,b,g

#plt.savefig("umutfoto.png") fotoyu kaydeder
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age vs weight")
plt.show()#we must do it to see data
```

```python
numpy_arr1 = np.linspace(0,10,20) # 10u 20 ye bolup arttirio
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1**3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2, "b-*")
plt.show()
```

```python
plt.plot(numpy_arr2,numpy_arr1, "b-*")
plt.show()
```

```python
plt.subplot(1,2,1) # bir row iki kolon 1. grafigi ciziyoruz
plt.plot(numpy_arr2,numpy_arr1, "r--")
plt.subplot(1,2,2) # bir row iki kolon 2. grafigi ciziyoruz
plt.plot(numpy_arr1,numpy_arr2, "r--")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.8,0.3]) # ilk 0.3 genislik ikinci 0.3 yukseklik,  0.1 ler ise kenar bosluklarini ayarliyor
figure_axes.plot(numpy_arr1,numpy_arr2, "g--")
figure_axes.set_xlabel("x")
figure_axes.set_ylabel("y")
figure_axes.set_title("title")
```

```python
from matplotlib.cbook import normalize_kwargs
new_fig = plt.figure(dpi= 100) # dimention
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1,numpy_arr1**2, label = "nump_array**2")
new_axes.plot(numpy_arr1,numpy_arr1**3, label = "nump_array**3")
new_axes.legend()# shows labels
#new_axes.legend(loc=1)--->changes positions of the label
plt.show()
```
